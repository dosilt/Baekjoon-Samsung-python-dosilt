```python
# N = int(input())
# matrix = [list(map(int, input().split(' ')))]

# N = 3
# matrix = [[0, 0, 0], [0, 0, 0], [0, 9, 0]]
## 0

# N = 3
# matrix = [[0, 0, 1], [0, 0, 0], [0, 9, 0]]
## 3

# N = 4
# matrix = [[4, 3, 2, 1], [0, 0, 0, 0], [0, 0, 9, 0], [1, 2, 3, 4]]
## 14

N = 6
matrix = [[5, 4, 3, 2, 3, 4], [4, 3, 2, 3, 4, 5], [3, 2, 9, 5, 6, 6], [2, 1, 2, 3, 4, 5], [3, 2, 1, 6, 5, 4], [6, 6, 6, 6, 6, 6]]
## 60

from collections import deque

for i in range(N):
    for j in range(N):
        if matrix[i][j] == 9:
            shark = deque([[i, j, 0]])
            matrix[i][j] = 0
shark_size = 2
shark_eat = 0
answer = 0
def bfs(shark, matrix, N, shark_size):
    dx = (-1, 0, 0, 1)
    dy = (0, -1, 1, 0)
    visited = [[False for _ in range(N)] for _ in range(N)]
    while shark:
        shark = sorted(shark, key=lambda x:[x[2], x[0]])
        x, y, dis = shark.popleft()
        visited[x][y] = True
        if 0 < matrix[x][y] < shark_size:
            matrix[x][y] = 0
            return x, y, dis
        
        for i in range(4):
            nx, ny = x+dx[i], y+dy[i]
            if 0<=nx<N and 0<=ny<N and matrix[nx][ny] <= shark_size and visited[nx][ny] == False:
                shark.append((nx, ny, dis+1))
                visited[nx][ny] = True
                
    return 0, 0, 0

while True:
    for aa in matrix:
        print(aa)
    print()
    loc_x, loc_y, dis = bfs(shark, matrix, N, shark_size)
    shark = deque([[loc_x, loc_y, 0]])
    shark_eat += 1
    if shark_size == shark_eat:
        shark_size += 1
        shark_eat = 0
    if dis == 0:
        break
    answer += dis
print(answer)
```
