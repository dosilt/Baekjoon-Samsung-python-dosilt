# https://www.acmicpc.net/problem/17143

```python
# row, col, sharks = map(int, input().split(' '))
# command = [list(map(int, input().split(' '))) for _ in range(sharks)]

row, col, sharks = 4, 6, 8
command = [[4, 1, 3, 3, 8], [1, 3, 5, 2, 9], [2, 4, 8, 4, 1], [4, 5, 0, 1, 4], [3, 3, 1, 2, 7], [1, 5, 8, 4, 3], [3, 6, 2, 1, 2], [2, 2, 2, 3, 5]]
## 22

# row, col, sharks = 100, 100, 0
# command = []
## 0

# row, col, sharks = 4, 5, 4
# command = [[4, 1, 3, 3, 8], [1, 3, 5, 2, 9], [2, 4, 8, 4, 1], [4, 5, 0, 1, 4]]
## 22

dn = {1:(-1, 0), 2:(1, 0), 3:(0, 1), 4:(0, -1)}
ch_dn = {1:2, 2:1, 3:4, 4:3}

matrix = [[[] for _ in range(col)] for _ in range(row)]

for r, c, s, d, z in command:
    if matrix[r-1][c-1]:
        if matrix[r-1][c-1][2] > z:
            continue
        else:
            matrix[r-1][c-1] = [s, d, z]
    else:
        matrix[r-1][c-1] = [s, d, z]

def catch(matrix, row, count, answer):
    for i in range(row):
        if matrix[i][count]:
            answer += matrix[i][count][2]
            matrix[i][count] = []
            return answer, matrix
    return answer, matrix

def next_position(row, col, direct, velocity, cur_x, cur_y):
    while True:
        if direct == 1: # 위를 바라봄
            if cur_x >= velocity:
                cur_x -= velocity
                return cur_x, cur_y, direct
            else:
                velocity -= cur_x
                cur_x = 0
                direct = 2

        elif direct == 2: # 아래를 바라봄
            if row-1-cur_x >= velocity:
                cur_x += velocity
                return cur_x, cur_y, direct
            else:
                velocity -= row-1-cur_x
                cur_x = row-1
                direct = 1
                
        elif direct == 3: # 오른쪽을 바라봄
            if col-1-cur_y >= velocity:
                cur_y += velocity
                return cur_x, cur_y, direct
            else:
                velocity -= col-1-cur_y
                cur_y = col-1
                direct = 4

        else: # 왼쪽을 바라봄
            if cur_y >= velocity:
                cur_y -= velocity
                return cur_x, cur_y, direct
            else:
                velocity -= cur_y
                cur_y = 0
                direct = 3
            
        

def move(matrix, row, col, dn, ch_dn):
    maps = [[[] for _ in range(col)] for _ in range(row)]
    for i in range(row):
        for j in range(col):
            if matrix[i][j]:
                nx, ny = i, j
                velocity, direct, size = matrix[i][j]
                nx, ny, direct = next_position(row, col, direct, velocity, nx, ny)
                if maps[nx][ny]:
                    if maps[nx][ny][2] > size:
                        continue
                    else:
                        maps[nx][ny] = [velocity, direct, size]
                else:
                    maps[nx][ny] = [velocity, direct, size]

    return maps

answer = 0

for j in range(col):
    answer, matrix = catch(matrix, row, j, answer)
    matrix = move(matrix, row, col, dn, ch_dn)
print(answer)
```

# 해설
시간초과와 많은 다툼이 있던 문제이다. python의 단점이 굉장히 두드러 지는 문제이다.

핵심 포인트만 집어서 서술 하자면 

a. 같은 위치에 상어가 2마리면 큰 거만 살아남는다.

즉 우리는 상어가 움직이기 전과 움직이기 후의 matrix를 따로 만들어야 할 필요성이 있다.

움직이지도 않은 상어가 잡아 먹힐 가능성이 존재하기 때문이다.

b. 벽에 부딫치면 180도 회전하여 돌아온다.

이 부분이 시간초과의 주범이다. 한칸씩 전진 후진하며 벽에 부딫쳤을 경우 회전을 넣으면 시간초과가 난다.

따라서 현재 위치에서 벽까지의 거리를 계산하여 벽보다 움직일 수 있는 거리가 가까우면 움직이고

벽에 닿는 거리면 벽까지만 움직이고 direct를 변경해 준다. 즉 한번에 벽 끝에서 끝으로 이동할 수 있으므로

많은 시간이 단축된다.
