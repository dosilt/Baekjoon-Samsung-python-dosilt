```python
from sys import exit

# col, count, row = list(map(int, input().split(' ')))
# matrix = [[0 for _ in range(col)] for _ in range(row)]
# command = [list(map(int, input().split(' '))) for i in range(count)]

# col, count, row = 5, 5, 6
# matrix = [[0 for _ in range(col)] for _ in range(row)]
# command = [[1, 1], [3, 2], [2, 3], [5, 1], [5, 4]]

# col, count, row = 2, 0, 3
# matrix = [[0 for _ in range(col)] for _ in range(row)]
# command = []

# col, count, row = 2, 1, 3
# matrix = [[0 for _ in range(col)] for _ in range(row)]
# command = [[1, 1]]


col, count, row = 6, 5, 6
matrix = [[0 for _ in range(col)] for _ in range(row)]
command = [[1, 1], [3, 2], [1, 3], [2, 5], [5, 5]]
answer = float('inf')
ladder = 1

for x, y in command:
    x -= 1
    y -= 1
    matrix[x][y] = ladder
    matrix[x][y+1] = ladder
    ladder += 1
def check(row, col, matrix):
    for j in range(col-1):
        loc = j
        for i in range(row):
            if matrix[i][loc] != 0:
                if 0<=loc-1 and matrix[i][loc-1] == matrix[i][loc]:
                    loc -= 1
                    continue
                if matrix[i][loc+1] == matrix[i][loc]:
                    loc += 1
        if j != loc:
            return False
    return True
        
def make_ladder(row, col, matrix, ladder, add, s_row):
    global answer
    
    if check(row, col, matrix):
        answer = min(answer, add)
        if answer == 0:
            exit()
        return
    
    if add == 3 or answer <= add:
        return
    
    for i in range(s_row, row):
        for j in range(col-1):
            if matrix[i][j] == 0 and matrix[i][j+1] == 0:
                matrix[i][j] = ladder
                matrix[i][j+1] = ladder
                ladder += 1
                make_ladder(row, col, matrix, ladder, add+1, i)
                matrix[i][j] = 0
                matrix[i][j+1] = 0
                ladder -= 1
                
if not check(row, col, matrix) or count != 0:
    make_ladder(row, col, matrix, ladder, 0, 0)
    if answer != float('inf'):
        print(answer)
    else:
        print(-1)
else:
    print(0)
```
