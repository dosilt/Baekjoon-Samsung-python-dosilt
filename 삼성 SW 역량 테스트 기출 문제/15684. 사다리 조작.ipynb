```python
N, bridges, count = list(map(int, input().split(' ')))
bridge = set()
for i in range(bridges):
    row, col = list(map(int, input().split(' ')))
    bridge.add((row, col, row, col+1))

def check(bridge, Row, Col):
    for j in range(1, Col+1):
        i = 1
        line = j
        while i < Row+1:
            if (i, line-1, i, line) in bridge:
                line = line-1
            elif (i, line, i, line+1) in bridge:
                line = line+1
            i += 1
        if line != j:
            return False
    return True

def make_bridge(count, bridge, Row, Col):
    global answer
    if check(bridge, Row, Col):
        answer = min(answer, count)
    
    if count == 3:
        return
    
    for i in range(1, Row+1):
        for j in range(1, Col):
            if (i, j, i, j+1) not in bridge and (i, j-1, i, j) not in bridge:
                bridge.add((i, j, i, j+1))
                make_bridge(count+1, bridge, Row, Col)
                bridge.remove((i, j, i, j+1))
                
count = 0
answer = float('inf')
make_bridge(count, bridge, Row, Col)
if answer == float('inf'):
    print(-1)
else:
    print(answer)
```
