```python
# N, low, high = list(map(int, input().split(' ')))
# matrix = [list(map(int, input().split(' '))) for i in range(N)]

# N, low, high = 2, 20, 50
# matrix = [[50, 30], [20, 40]]

# N, low, high = 2, 40, 50
# matrix = [[50, 30], [20, 40]]

N, low, high = 3, 5, 10
matrix = [[10, 15, 20], [20, 30, 25], [40, 22, 10]]

from collections import defaultdict

def check(N, matrix, low, high):
    answer = 0

    while True:
        maps = [[float('inf') for _ in range(N)] for _ in range(N)]
        dx = (-1, 0)
        dy = (0, -1)
        count_ = 2
        maps[0][0] = 1
        loc = defaultdict(list)
        sum_ = defaultdict(int)
        count = defaultdict(int)

        loc[1].append((0, 0))
        sum_[1] += matrix[0][0]    
        count[1] += 1
        for i in range(N):
            for j in range(N):
                if i == 0 and j == 0:
                    continue

                for n in range(2):
                    if 0 <= i+dx[n] < N and 0 <= j+dy[n] < N and low <= abs(matrix[i][j] - matrix[i+dx[n]][j+dy[n]]) <= high:
                        maps[i][j] = maps[i+dx[n]][j+dy[n]] = min(maps[i+dx[n]][j+dy[n]], maps[i][j])
                        
                        loc[maps[i+dx[n]][j+dy[n]]].append((i, j))
                        sum_[maps[i+dx[n]][j+dy[n]]] += matrix[i][j]
                        count[maps[i+dx[n]][j+dy[n]]] += 1
                print(i, j, maps[i][j])
                if maps[i][j] == float('inf'):
                    maps[i][j] = count_
                    loc[count_].append((i, j))
                    sum_[count_] += matrix[i][j]
                    count[count_] += 1
                    count_ += 1
                    
        for aa in maps:
            print(aa)
        print(loc)
        print(sum_)
        print(count)
        
        if count_ == N*N+1:
            return answer
            
        for n in loc:
            data = sum_[n] // count[n]
            if len(loc[n]) > 1:
                for i, j in loc[n]:
                    matrix[i][j] = data
        answer += 1
check(N, matrix, low, high)
```
